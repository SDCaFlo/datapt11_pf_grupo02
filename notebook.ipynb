```python
import pandas as pd

# Cargar el dataset
ruta_csv = "HawaiiMuestra.csv"  # Asegúrate de colocar la ruta correcta
df = pd.read_csv(ruta_csv)

# Contar la cantidad de veces que aparece cada nombre en 'name_x'
frecuencia_nombres = df['name_x'].value_counts()

# Filtrar nombres que se repiten más de 6 veces
nombres_mas_de_6 = frecuencia_nombres[frecuencia_nombres > 6]

# Mostrar resultados
print("Nombres que se repiten más de 6 veces en la columna 'name_x':\n")
print(nombres_mas_de_6)

# Mostrar el total de nombres que cumplen con la condición
print("\nTotal de nombres que se repiten más de 6 veces:", nombres_mas_de_6.shape[0])

```
